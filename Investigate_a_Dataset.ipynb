### General Properties

```python
# Load your data and print out a few lines. Perform operations to inspect data
#   types and look for instances of missing or possibly errant data.

df1 = pan.read_csv('tmdb-movies.csv') 

df1.head()


```



### Data Cleaning (Replace this with more specific notes!)

Data Cleaning
Remvoe the outliers as the following:

 1- Remove all movies where the budget is less than 10000 USD

 2- Remove all movies where the revenue is less than 10000 USD
 
 3- Remove all movies where its duration runtime less than 30 minutes

Split the keywords column to get the first keyword

Split the geners column to get the first genre



```python
# Remove all movies where the budget is less than 10000 USD
df_index=df1[df1['budget']<10000].index
df1.drop(df_index , inplace=True)


# Remove all movies where the revenue is less than 10000 USD
df_index=df1[df1['revenue']<10000].index
df1.drop(df_index , inplace=True)


# Remove all movies where its duration runtime less than 30 minutes
df_index=df1[df1['runtime']<30].index
df1.drop(df_index , inplace=True)

#Split the keywords column to get the first keyword
df1['keywords']=df1['keywords'].str.split('|',n=1,expand=True)

# Split the genres column to get the first genre
df1['genres']=df1['genres'].str.split('|',n=1,expand=True)

df1.head()
```

## Exploratory Data Analysis



### Research Question 1 (Replace this header name!)

Question 1
What is the movies genre which has the highest revenue? And what is the distribution of highest revenue genre?

To answer this question I will calculate the revenue of movie grouped by genre.

```python
# Use this, and more code cells, to explore your data. Don't forget to add
#   Markdown cells to document your observations and findings.

values=df1.groupby('genres')['revenue'].sum()
locations=values.index
mp.figure(figsize=(20,10))
mp.bar(locations,values)
mp.title('Revenue of Movies Genres')
mp.xlabel('Movies Genres')
mp.ylabel('Revenue')
mp.show()
```

As per the chat above, i found that the Action genre is the highest revenue compared to other geners,
I will create new data frame in order to include reocords of Action genre only, then I will plot depending on the release year using histogram chart to represent the distribution.


```python
df2=df1[df1['genres']=='Action']
mp.figure(figsize=(20,10))
mp.hist(df2['release_year'],bins='auto')
mp.title('Action Movie per Year')
mp.xlabel('Release Year')
mp.ylabel('Count of Movies')
```

> **The analysis of question1:**

>The first chart: I see that revenue of action movies is higher than other genres in the history of movies as per the imdb data, this is an indicator that the people prefer watching the action movie more than other genres.

>The second chart: I see that the revenue of Action Movies was low before 1990 but after 2010 Year I see there is jump in the revenue which is an indicator that watchers become prefer action movies so the producing companies doubled their investements  in this genre.
